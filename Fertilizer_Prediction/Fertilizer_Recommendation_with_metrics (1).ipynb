```python
def plot_conti(x):
    fig, axes = plt.subplots(nrows=1,ncols=3,figsize=(15,5),tight_layout=True)
    axes[0].set_title('Histogram')
    sns.histplot(x,ax=axes[0])
    axes[1].set_title('Checking Outliers')
    sns.boxplot(x,ax=axes[1])
    axes[2].set_title('Relation with output variable')
    sns.boxplot(y = x,x = data.Fertilizer)

def plot_cato(x):
    fig, axes = plt.subplots(nrows=1,ncols=2,figsize=(15,5),tight_layout=True)
    axes[0].set_title('Count Plot')
    sns.countplot(x,ax=axes[0])
    axes[1].set_title('Relation with output variable')
    sns.countplot(x = x,hue = data.Fertilizer, ax=axes[1])
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
encode_soil = LabelEncoder()

data.Soil_color = encode_soil.fit_transform(data.Soil_color)

Soil_Type = pd.DataFrame(zip(encode_soil.classes_,encode_soil.transform(encode_soil.classes_)),columns=['Original','Encoded'])
Soil_Type = Soil_Type.set_index('Original')
Soil_Type
```

```python
encode_crop = LabelEncoder()

data.Crop = encode_crop.fit_transform(data.Crop)

Crop_Type = pd.DataFrame(zip(encode_crop.classes_,encode_crop.transform(encode_crop.classes_)),columns=['Original','Encoded'])
Crop_Type = Crop_Type.set_index('Original')
Crop_Type
```

```python
encode_ferti = LabelEncoder()


data.Fertilizer = encode_ferti.fit_transform(data.Fertilizer)


Fertilizer = pd.DataFrame(zip(encode_ferti.classes_,encode_ferti.transform(encode_ferti.classes_)),columns=['Original','Encoded'])
Fertilizer = Fertilizer.set_index('Original')
Fertilizer
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(data.drop('Fertilizer',axis=1),data.Fertilizer,test_size=0.2,random_state=1)
print('Shape of Splitting :')
print('x_train = {}, y_train = {}, x_test = {}, y_test = {}'.format(x_train.shape,y_train.shape,x_test.shape,y_test.shape))
```

```python
x_train.info()
```

```python

import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import fbeta_score, r2_score, precision_score, classification_report


```

```python

model_scores = {
    "RandomForest": {},
    "RandomForest (GridSearch)": {},
    "DecisionTree": {}
}

```

```python
# RandomForestClassifier without GridSearch
rand = RandomForestClassifier()
pred_rand = rand.fit(x_train, y_train).predict(x_test)

# Calculating metrics for RandomForest
model_scores["RandomForest"]["F2 Score"] = fbeta_score(y_test, pred_rand, beta=2, average='weighted')
model_scores["RandomForest"]["Weighted Precision"] = precision_score(y_test, pred_rand, average='weighted')
try:
    model_scores["RandomForest"]["R2 Score"] = r2_score(y_test, pred_rand)
except:
    model_scores["RandomForest"]["R2 Score"] = None
```

```python
params = {
    'n_estimators': [300, 400, 500],
    'max_depth': [5, 10, 15],
    'min_samples_split': [2, 5, 8]
}

# GridSearchCV with RandomForestClassifier
grid_rand = GridSearchCV(RandomForestClassifier(), params, cv=3, verbose=3, n_jobs=-1)
grid_rand.fit(x_train, y_train)
pred_grid_rand = grid_rand.predict(x_test)

# Calculate metrics for RandomForest with GridSearchCV
model_scores["RandomForest (GridSearch)"]["F2 Score"] = fbeta_score(y_test, pred_grid_rand, beta=2, average='weighted')
model_scores["RandomForest (GridSearch)"]["Weighted Precision"] = precision_score(y_test, pred_grid_rand, average='weighted')
try:
    model_scores["RandomForest (GridSearch)"]["R2 Score"] = r2_score(y_test, pred_grid_rand)
except:
    model_scores["RandomForest (GridSearch)"]["R2 Score"] = None
```

```python

# DecisionTreeClassifier
dt = DecisionTreeClassifier(random_state=42)
pred_dec = dt.fit(x_train, y_train).predict(x_test)

# Calculate metrics for DecisionTree
model_scores["DecisionTree"]["F2 Score"] = fbeta_score(y_test, pred_dec, beta=2, average='weighted')
model_scores["DecisionTree"]["Weighted Precision"] = precision_score(y_test, pred_dec, average='weighted')
try:
    model_scores["DecisionTree"]["R2 Score"] = r2_score(y_test, pred_dec)
except:
    model_scores["DecisionTree"]["R2 Score"] = None
```

```python
# Prepare data for plotting
metrics = ["F2 Score", "Weighted Precision", "R2 Score"]
models = list(model_scores.keys())

# Extracting score values for each metric by model
f2_scores = [model_scores[model]["F2 Score"] for model in models]
precision_scores = [model_scores[model]["Weighted Precision"] for model in models]
r2_scores = [model_scores[model]["R2 Score"] if model_scores[model]["R2 Score"] is not None else 0 for model in models]
```

```python
from sklearn.metrics import accuracy_score

# Calculate accuracy for RandomForest without GridSearch
accuracy_rand = accuracy_score(y_test, pred_rand)
model_scores["RandomForest"]["Accuracy"] = accuracy_rand

# Calculate accuracy for RandomForest with GridSearch
accuracy_grid_rand = accuracy_score(y_test, pred_grid_rand)
model_scores["RandomForest (GridSearch)"]["Accuracy"] = accuracy_grid_rand

# Calculate accuracy for DecisionTree
accuracy_dec = accuracy_score(y_test, pred_dec)
model_scores["DecisionTree"]["Accuracy"] = accuracy_dec

# Prepare data for printing in tabular form
model_scores_df = pd.DataFrame(model_scores).T  # Transpose for better readability
print(model_scores_df)

```

```python

x = range(len(models))
width = 0.2

plt.figure(figsize=(10, 6))
plt.bar([pos - width for pos in x], f2_scores, width, label="F2 Score", color='lightcoral')
plt.bar(x, precision_scores, width, label="Weighted Precision", color='lightblue')
plt.bar([pos + width for pos in x], r2_scores, width, label="R2 Score", color='lightgreen')

plt.xlabel("Models")
plt.ylabel("Score")
plt.title("Comparative Analysis of Model Scores")
plt.xticks(x, models)
plt.ylim(0, 1)
plt.legend()
plt.tight_layout()
plt.show()
```
